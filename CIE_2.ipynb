```python
import numpy as np
from scipy import stats
mean1 = np.mean(data['Age'])
median1 = np.median(data['Age'])
mode1 = stats.mode(data['Age'])
mean2 = np.mean(data['Fare'])
median2 = np.median(data['Fare'])
mode2 = stats.mode(data['Fare'])
print("Central Trendency:")
print("Mean Age:", mean1)
print("Mode Age:", mode1)
print("Median Age", median1)
print("mean Fare:", mean2)
print("Mode Fare:", mode2)
print("Median Fare:", median2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data.replace(to_replace='survived', value=1, inplace=True)
data.replace(to_replace='not survived', value=0, inplace=True)
plt.figure(figsize=(5,5))
data['Survived'].value_counts().plot(kind= "bar", color=["red", "blue"])
plt.title("survival status")
plt.xlabel('status')
plt.ylabel('count')
plt.xticks(ticks=[0,1], labels=['not survived', 'survived'], rotation=0)
plt.show()
```
